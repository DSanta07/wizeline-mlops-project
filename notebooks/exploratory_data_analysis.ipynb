```python
import sys
sys.path.append("./../projectname")
from projectname.config import raw_data_dir
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

# Overview

```python
hotel_bookings_path = raw_data_dir / "hotel_bookings.csv"
hotel_bookings_df = pd.read_csv(hotel_bookings_path)
hotel_bookings_df
```

```python
# Display the first few rows of the dataset
hotel_bookings_df.head()
```

# Summary Statistics

```python
# Summary statistics
summary_stats = hotel_bookings_df.describe(include='all')
summary_stats
```

# Key Variables

```python
# Distribution of key variables
key_variables = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies', 'adr']
plt.figure(figsize=(14, 10))
for i, var in enumerate(key_variables, 1):
    plt.subplot(3, 3, i)
    sns.histplot(hotel_bookings_df[var], bins=30, kde=True)
    plt.title(f'Distribution of {var}')
plt.tight_layout()
plt.show()
```

# Booking Cancellation Analysis

```python
# Cancellation rate
cancellation_rate = hotel_bookings_df['is_canceled'].mean()
cancellation_rate

# Plot overall cancellation rate
plt.subplot(1, 2, 1)
sns.barplot(x=['Overall'], y=[cancellation_rate])
plt.ylim(0, 1)
plt.ylabel('Cancellation Rate')
plt.title('Overall Cancellation Rate')
```

```python
# Analyze cancellation by hotel type
cancellation_by_hotel = hotel_bookings_df.groupby('hotel')['is_canceled'].mean()
cancellation_by_hotel

# Plot cancellation rate by hotel type
plt.subplot(1, 2, 2)
sns.barplot(x=cancellation_by_hotel.index, y=cancellation_by_hotel.values)
plt.ylim(0, 1)
plt.ylabel('Cancellation Rate')
plt.title('Cancellation Rate by Hotel Type')

plt.tight_layout()
plt.show()
```

# Booking Trends

```python
# Convert reservation_status_date to datetime
hotel_bookings_df['reservation_status_date'] = pd.to_datetime(hotel_bookings_df['reservation_status_date'])

# Add year and month columns for analysis
hotel_bookings_df['year'] = hotel_bookings_df['reservation_status_date'].dt.year
hotel_bookings_df['month'] = hotel_bookings_df['reservation_status_date'].dt.month

# Booking trends over time
bookings_per_year = hotel_bookings_df['year'].value_counts().sort_index()
bookings_per_month = hotel_bookings_df['month'].value_counts().sort_index()

# ADR trends over time
adr_per_month_year = hotel_bookings_df.groupby(['year', 'month'])['adr'].mean().unstack()
```

```python
# Plot booking trends
plt.figure(figsize=(14, 8))

# Bookings per year
plt.subplot(2, 1, 1)
sns.lineplot(x=bookings_per_year.index, y=bookings_per_year.values, marker='o')
plt.ylabel('Number of Bookings')
plt.title('Number of Bookings per Year')
```

```python
# Bookings per month
plt.subplot(2, 1, 2)
sns.lineplot(x=bookings_per_month.index, y=bookings_per_month.values, marker='o')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.ylabel('Number of Bookings')
plt.title('Number of Bookings per Month')

plt.tight_layout()
plt.show()
```

```python
# Plot ADR trends
plt.figure(figsize=(14, 6))
sns.heatmap(adr_per_month_year, annot=True, fmt=".2f", cmap="YlGnBu", cbar_kws={'label': 'ADR'})
plt.ylabel('Year')
plt.xlabel('Month')
plt.title('Average Daily Rate (ADR) per Month and Year')
plt.show()
```
